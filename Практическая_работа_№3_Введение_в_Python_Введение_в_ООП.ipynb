## **Задание №1. Создайте класс Sphere для представления сферы в трехмерном пространстве**


Обеспечьте следующие методы класса:

1. Конструктор, принимающий 4 действительных числа: радиус, и 3 координаты центра шара. Если конструктор вызывается без аргументов, создать объект сферы с единичным радиусом и центром в начале координат. Если конструктор вызывается с аргументом, создать объект сферы с соответствующим радиусом и центром в начале координат.

2. Метод get_volume (), который возвращает действительное число — объем шара, ограниченной текущей сферой.

3. Метод get_square (), который возвращает действительное число — площадь внешней поверхности сферы.

4. Метод get_radius (), который возвращает действительное число — радиус сферы.

5. Метод get_center (), который возвращает тьюпл с 3 действительными числами — координатами центра сферы в том же порядке, в каком они задаются в конструкторе.

6. Метод set_radius (r), который принимает 1 аргумент — действительное число, и меняет радиус текущей сферы, ничего не возвращая.

7. Метод set_center (x, y, z), который принимает 3 аргумента — действительных числа, и меняет координаты центра сферы, ничего не возвращая. Координаты задаются в том же порядке, что и в конструкторе.

8. Метод is_point_inside (x, y, z), который принимает 3 аргумента — действительных числа — координаты некоторой точки в пространстве (в том же порядке, что и в конструкторе), и возвращает логическое значение True или False в зависимости от того, находится эта точка внутри сферы.

````python
from math import pi, sqrt
class Sphere:
  def __init__(self, r=1., x=0., y=0., z=0.,):
    self._x = float(x)
    self._y = float(y)
    self._z = float(z)
    self._r = float(r)

  def get_volume(self):
    return 4/3*pi*self._r**3

  def get_square(self):
    return 4*pi*self._r**2

  def get_radius(self):
    return self._r

  def get_center(self):
    return self._x, self._y, self._z

  def set_radius(self, r):
    self._r = r

  def set_center(self, x, y, z):
    self._x = float(x)
    self._y = float(y)
    self._z = float(z)
  def is_point_inside(self, x, y, z):
    d = sqrt((x-self._x)**2+(y-self._y)**2+(z-self._z)**2)
    if d < self._r:
      return True
    return False
````

````python
# Тесты

s0 = Sphere(0.5) # test sphere creation with radius and default center
print(s0.get_center()) # (0.0, 0.0, 0.0)
print(s0.get_volume()) # 0.523598775598
print(s0.is_point_inside(0 , -1.5, 0)) # False
s0.set_radius(1.6)
print(s0.is_point_inside(0, -1.5, 0)) # True
print(s0.get_radius()) # 1.6
````

## **Задание №2. Создайте класс SuperStr, который наследует стандартный класс str и содержит 2 новых метода:**

1. Метод is_repeatance (s), который принимает 1 аргумент s и возвращает True или False в зависимости от того, может ли текущая строку быть получена целым количеством повторов строки s. Вернуть False, если s не является строкой. Считать, что пустая строка не содержит повторов.

2. Метод is_palindrom (), который возвращает True или False в зависимости от того, является ли строка [палиндромом](https://ru.wikipedia.org/wiki/Палиндром). Регистрами символов пренебрегать. Пустую строку считать палиндромом.

````python
class SuperStr(str):
    def is_repeatance(self, s):
        if not isinstance(s, str):
            return False
        if len(self) % len(s) != 0:
            return False
        repeat_count = len(self) // len(s)
        return self == s * repeat_count

    def is_palindrom(self):
        if not self:
            return True
        cleaned = self.lower().replace(' ', '')
        return cleaned == cleaned[::-1]
````

````python
# Тесты

s = SuperStr("123123123123")
print(s.is_repeatance("123")) # True
print(s.is_repeatance("123123")) # True
print(s.is_repeatance("123123123123")) # True
print(s.is_repeatance("12312")) # False
print(s.is_repeatance(123)) # False
print(s.is_palindrom()) # False
print(s) # 123123123123 (строка)
print(int(s)) # 123123123123 (целое число)
print(s + "qwe") # 123123123123qwe
p = SuperStr("123_321")
print(p.is_palindrom()) # True
````

````python
class Student:
    def __init__(self, name, surname, gender):
        self.name = name
        self.surname = surname
        self.gender = gender
        self.finished_courses = []
        self.courses_in_progress = []
        self.grades = {}

    def add_courses(self, course_name):
        self.finished_courses.append(course_name)

    def add_lecture_rate(self, course, lecturer, grade):
        if course not in lecturer.lecture_rate:
            lecturer.lecture_rate[course] = []
        lecturer.lecture_rate[course].append(grade)

    def __str__(self):
        # Вычисляем среднюю оценку за домашние задания
        avg_grade = 0
        if self.grades:
            all_grades = []
            for course_grades in self.grades.values():
                # Проверяем, является ли course_grades списком или числом
                if isinstance(course_grades, list):
                    all_grades.extend(course_grades)
                else:
                    all_grades.append(course_grades)
            if all_grades:
                avg_grade = sum(all_grades) / len(all_grades)

        # Форматируем курсы
        courses_in_progress_str = ', '.join(self.courses_in_progress) if self.courses_in_progress else "Нет"
        finished_courses_str = ', '.join(self.finished_courses) if self.finished_courses else "Нет"

        return f"Имя: {self.name}\nФамилия: {self.surname}\n" \
               f"Средняя оценка за домашние задания: {avg_grade:.1f}\n" \
               f"Курсы в процессе изучения: {courses_in_progress_str}\n" \
               f"Завершенные курсы: {finished_courses_str}"

    def __eq__(self, other):
        """Сравниваем студентов по средней оценке за домашние задания."""
        if not isinstance(other, Student):
            return NotImplemented

        # Вычисляем среднюю оценку текущего студента
        avg_self = 0
        if self.grades:
            all_grades = []
            for course_grades in self.grades.values():
                if isinstance(course_grades, list):
                    all_grades.extend(course_grades)
                else:
                    all_grades.append(course_grades)
            if all_grades:
                avg_self = sum(all_grades) / len(all_grades)

        # Вычисляем среднюю оценку другого студента
        avg_other = 0
        if other.grades:
            all_grades = []
            for course_grades in other.grades.values():
                if isinstance(course_grades, list):
                    all_grades.extend(course_grades)
                else:
                    all_grades.append(course_grades)
            if all_grades:
                avg_other = sum(all_grades) / len(all_grades)

        return avg_self == avg_other

    def __lt__(self, other):
        """Сравниваем, у кого средняя оценка меньше."""
        if not isinstance(other, Student):
            return NotImplemented

        # Вычисляем среднюю оценку текущего студента
        avg_self = self._calculate_average_grade()
        avg_other = other._calculate_average_grade()

        return avg_self < avg_other

    def __gt__(self, other):
        """Сравниваем, у кого средняя оценка больше."""
        if not isinstance(other, Student):
            return NotImplemented

        avg_self = self._calculate_average_grade()
        avg_other = other._calculate_average_grade()

        return avg_self > avg_other

    def _calculate_average_grade(self):
        """Вспомогательный метод для расчета средней оценки."""
        avg = 0
        if self.grades:
            all_grades = []
            for course_grades in self.grades.values():
                if isinstance(course_grades, list):
                    all_grades.extend(course_grades)
                else:
                    all_grades.append(course_grades)
            if all_grades:
                avg = sum(all_grades) / len(all_grades)
        return avg


class Mentor:
    def __init__(self, name, surname):
        self.name = name
        self.surname = surname
        self.courses_attached = []

    def rate_hw(self, student, course, grade):
        student.grades[course] = grade  # Перезаписываем оценку

    def __str__(self):
        return f"Имя: {self.name}\nФамилия: {self.surname}"


class Lecturer(Mentor):
    def __init__(self, name, surname):
        super().__init__(name, surname)
        self.lecture_rate = {}  # курс -> список оценок

    def rate_hw(self, student, course, grade):
        raise NotImplementedError("Лектор не может ставить оценки")

    def __str__(self):
        # Вычисляем среднюю оценку за лекции
        avg_rate = 0
        if self.lecture_rate:
            all_rates = []
            for course_rates in self.lecture_rate.values():
                all_rates.extend(course_rates)
            if all_rates:
                avg_rate = sum(all_rates) / len(all_rates)

        return f"Имя: {self.name}\nФамилия: {self.surname}\n" \
               f"Средняя оценка за лекции: {avg_rate:.1f}"

    def __eq__(self, other):
        """Сравниваем лекторов по средней оценке за лекции."""
        if not isinstance(other, Lecturer):
            return NotImplemented

        # Вычисляем среднюю оценку текущего лектора
        avg_self = 0
        if self.lecture_rate:
            all_rates = []
            for course_rates in self.lecture_rate.values():
                all_rates.extend(course_rates)
            if all_rates:
                avg_self = sum(all_rates) / len(all_rates)

        # Вычисляем среднюю оценку другого лектора
        avg_other = 0
        if other.lecture_rate:
            all_rates = []
            for course_rates in other.lecture_rate.values():
                all_rates.extend(course_rates)
            if all_rates:
                avg_other = sum(all_rates) / len(all_rates)

        return avg_self == avg_other

    def __lt__(self, other):
        """Сравниваем, у какого лектора средняя оценка меньше."""
        if not isinstance(other, Lecturer):
            return NotImplemented

        avg_self = self._calculate_average_rate()
        avg_other = other._calculate_average_rate()

        return avg_self < avg_other

    def __gt__(self, other):
        """Сравниваем, у какого лектора средняя оценка больше."""
        if not isinstance(other, Lecturer):
            return NotImplemented

        avg_self = self._calculate_average_rate()
        avg_other = other._calculate_average_rate()

        return avg_self > avg_other

    def _calculate_average_rate(self):
        """Вспомогательный метод для расчета средней оценки за лекции."""
        avg = 0
        if self.lecture_rate:
            all_rates = []
            for course_rates in self.lecture_rate.values():
                all_rates.extend(course_rates)
            if all_rates:
                avg = sum(all_rates) / len(all_rates)
        return avg


class Reviewer(Mentor):
    def rate_hw(self, student, course, grade, lecturer):
        # Ставим оценку студенту (перезаписываем)
        student.grades[course] = grade

        # Студент автоматически ставит оценку лектору
        student.add_lecture_rate(course, lecturer, grade)

    def __str__(self):
        # Используем родительский метод, так как формат такой же
        return super().__str__()
````

````python
import math

class Fraction:
    def __init__(self, numerator: int, denominator: int):
        if denominator == 0:
            raise ValueError("Знаменатель не может быть нулем")
        self._numerator = numerator
        self._denominator = denominator
        self._simplify()

    def _simplify(self):
        """Приводит дробь к несократимому виду и нормализует знак."""
        gcd = math.gcd(abs(self._numerator), abs(self._denominator))
        self._numerator //= gcd
        self._denominator //= gcd

        # Переносим минус в числитель, чтобы знаменатель был положительным
        if self._denominator < 0:
            self._numerator = -self._numerator
            self._denominator = -self._denominator

    def __add__(self, other):
        if not isinstance(other, Fraction):
            return NotImplemented
        num = self._numerator * other._denominator + other._numerator * self._denominator
        den = self._denominator * other._denominator
        return Fraction(num, den)

    def __sub__(self, other):
        if not isinstance(other, Fraction):
            return NotImplemented
        num = self._numerator * other._denominator - other._numerator * self._denominator
        den = self._denominator * other._denominator
        return Fraction(num, den)

    def __mul__(self, other):
        if not isinstance(other, Fraction):
            return NotImplemented
        num = self._numerator * other._numerator
        den = self._denominator * other._denominator
        return Fraction(num, den)

    def __truediv__(self, other):
        if not isinstance(other, Fraction):
            return NotImplemented
        if other._numerator == 0:
            print("Деление на ноль")
            raise ZeroDivisionError("Деление на ноль")
        num = self._numerator * other._denominator
        den = self._denominator * other._numerator
        return Fraction(num, den)

    def __eq__(self, other):
        if not isinstance(other, Fraction):
            return NotImplemented
        return self._numerator == other._numerator and self._denominator == other._denominator

    def __lt__(self, other):
        if not isinstance(other, Fraction):
            return NotImplemented
        # Сравниваем крест-накрест: a/b < c/d  <=>  a*d < c*b (при положительных знаменателях)
        return self._numerator * other._denominator < other._numerator * self._denominator

    def __le__(self, other):
        return self < other or self == other

    def __gt__(self, other):
        return not self <= other

    def __ge__(self, other):
        return not self < other

    def __float__(self):
        return self._numerator / self._denominator

    def reciprocal(self):
        if self._numerator == 0:
            print("Обратная дробь к нулю не существует")
            raise ZeroDivisionError("Обратная дробь к нулю не существует")
        return Fraction(self._denominator, self._numerator)

    def __str__(self):
        if self._denominator == 1:
            return str(self._numerator)
        return f"{self._numerator}/{self._denominator}"

    def __repr__(self):
        return f"Fraction({self._numerator}, {self._denominator})"
    @classmethod
    def from_float(cls, value: float):
        if not isinstance(value, float):
            raise TypeError("Аргумент должен быть числом с плавающей точкой")
        if value == 0.0:
            return cls(0, 1)
        # Получаем точное представление float как дробь
        numerator, denominator = value.as_integer_ratio()
        return cls(numerator, denominator)
````
