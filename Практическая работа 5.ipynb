### **Содержимое модуля transformations.py:**

```python
%%writefile geo_transform/transformations.py
from math import sqrt, atan2, sin, cos
def cartesian_to_spherical(x,y,z):
  r = sqrt(x**2+y**2+z**2)
  azimuth = 2* atan2(y,x+sqrt(x**2+y**2))
  angle = atan2(sqrt(x**2+y**2),z)
  return r, azimuth, angle

def spherical_to_cartesian(r, azimuth, angle):
  x = r*sin(angle)*cos(azimuth)
  y = r*sin(angle)*sin(azimuth)
  z = r*cos(angle)
  return x, y, z
```

### **Содержимое модуля coordinates.py:**

```python
%%writefile geo_transform_oop/coordinates.py
from math import sqrt, atan2, sin, cos, pi

class CartesianPoint:
    def __init__(self, x: float, y: float, z: float):
        self.x = x
        self.y = y
        self.z = z

    def to_spherical(self) -> "SphericalPoint":
        r = sqrt(self.x**2 + self.y**2 + self.z**2)
        if r == 0:
            return SphericalPoint(0, 0, 0)
        azimuth = atan2(self.y, self.x)  # стандартное определение
        polar_angle = atan2(sqrt(self.x**2 + self.y**2), self.z)
        return SphericalPoint(r, azimuth, polar_angle)

    def __repr__(self):
        return f"CartesianPoint(x={self.x}, y={self.y}, z={self.z})"

    def __str__(self):
        return f"{self.x} {self.y} {self.z}"


class SphericalPoint:
    def __init__(self, r: float, azimuth: float, polar_angle: float):
        if r < 0:
            raise ValueError("Радиус не может быть отрицательным")
        self.r = r
        self.azimuth = azimuth          # угол в плоскости XY (от оси X)
        self.polar_angle = polar_angle  # угол от оси Z (0 — северный полюс)

    def to_cartesian(self) -> "CartesianPoint":
        x = self.r * sin(self.polar_angle) * cos(self.azimuth)
        y = self.r * sin(self.polar_angle) * sin(self.azimuth)
        z = self.r * cos(self.polar_angle)
        return CartesianPoint(x, y, z)

    def __repr__(self):
        return f"SphericalPoint(r={self.r}, azimuth={self.azimuth}, polar_angle={self.polar_angle})"

    def __str__(self):
        return f"{self.r} {self.azimuth} {self.polar_angle}"
```

### **Содержимое модуля converters.py:**

```python
%%writefile geo_transform_oop/converters.py
from .coordinates import CartesianPoint, SphericalPoint

class CoordinateConverter:
    @staticmethod
    def cartesian_to_spherical(x, y, z) -> SphericalPoint:
        return CartesianPoint(x, y, z).to_spherical()

    @staticmethod
    def spherical_to_cartesian(r, azimuth, polar_angle) -> CartesianPoint:
        return SphericalPoint(r, azimuth, polar_angle).to_cartesian()
```

### **Содержимое модуля utils.py:**

```python
%%writefile geo_transform_oop/utils.py
from math import pi

def deg_to_rad(deg: float) -> float:
    return deg * pi / 180.0

def rad_to_deg(rad: float) -> float:
    return rad * 180.0 / pi
```

### **Содержимое модуля file_handler.py:**

```python
%%writefile geo_transform_oop/file_handler.py
class FileHandler:
    def __init__(self, default_path: str = "result.txt"):
        self.default_path = default_path

    def write_result(self, data, path: str = None):
        path = path or self.default_path
        with open(path, mode='w', encoding='utf-8') as file:
            file.write(str(data))

    def read_coordinates(self, path: str) -> str:
        with open(path, mode='r', encoding='utf-8') as file:
            return file.read().strip()
```

### **Содержимое модуля geo_transform/\_\_main\_\_.py

```python
%%writefile geo_transform_oop/__main__.py
from .coordinates import CartesianPoint, SphericalPoint
from .utils import deg_to_rad, rad_to_deg
from .file_handler import FileHandler
import sys

def main():
    result = None
    file_handler = FileHandler()

    print("Добро пожаловать в GeoTransform (OOP-версия)!")
    while True:
        print("\n" + "="*50)
        print("Текущий результат:", result)
        print("""
1) Декартовы → Сферические
2) Сферические → Декартовы
3) Градусы → Радианы
4) Радианы → Градусы
5) Сохранить результат в файл
6) Загрузить координаты из файла
0) Выход
        """)
        choice = input("Выберите действие: ").strip()

        if choice == "1":
            try:
                coords = list(map(float, input("Введите x y z через пробел: ").split()))
                if len(coords) != 3:
                    raise ValueError
                point = CartesianPoint(*coords)
                result = point.to_spherical()
            except Exception as e:
                print("Ошибка ввода:", e)

        elif choice == "2":
            try:
                coords = list(map(float, input("Введите r azimuth polar_angle через пробел: ").split()))
                if len(coords) != 3:
                    raise ValueError
                point = SphericalPoint(*coords)
                result = point.to_cartesian()
            except Exception as e:
                print("Ошибка ввода:", e)

        elif choice == "3":
            try:
                deg = float(input("Введите градусы: "))
                result = deg_to_rad(deg)
            except Exception as e:
                print("Ошибка:", e)

        elif choice == "4":
            try:
                rad = float(input("Введите радианы: "))
                result = rad_to_deg(rad)
            except Exception as e:
                print("Ошибка:", e)

        elif choice == "5":
            if result is None:
                print("Нет данных для сохранения!")
            else:
                path = input("Путь к файлу (Enter — result.txt): ").strip() or "result.txt"
                file_handler.write_result(result, path)
                print(f"Результат сохранён в {path}")

        elif choice == "6":
            try:
                path = input("Путь к файлу: ").strip()
                content = file_handler.read_coordinates(path)
                print(f"Содержимое файла: {content}")
                # Попробуем автоматически распарсить?
                numbers = list(map(float, content.split()))
                if len(numbers) == 3:
                    print("Обнаружено 3 числа — можно использовать в преобразованиях")
                result = content  # или numbers?
            except Exception as e:
                print("Ошибка чтения файла:", e)

        elif choice == "0":
            print("Выход...")
            break

        else:
            print("Неверный выбор. Попробуйте снова.")

if __name__ == "__main__":
    main()
```
